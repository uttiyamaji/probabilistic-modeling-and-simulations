## Mandelbrot sets with python

```python
def mandel(c, maxiter):
    z = complex(0,0)
    
    for i in range(maxiter):
        z = (z*z)+c
        if abs(z) > 4:
            break
            pass
        pass
    return i
```

```python
mandel(complex(0.5,0.5), 40)
```

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def mandelbrot( maxiter, x_min, x_max, y_min, y_max, ns):
    x_vals = np.linspace(x_min, x_max, ns)
    y_vals = np.linspace(y_min, y_max, ns)
    
    xlen = len(x_vals)
    ylen = len(y_vals)

    atlas = np.zeros((xlen,ylen))

    for ix in range(xlen):
        for iy in range(ylen):
        
            cx = x_vals[ix]
            cy = y_vals[iy]
        
            c = complex(cx,cy)
        
            atlas[ix,iy] = mandel(c, maxiter)
            pass
        pass

    plt.figure(figsize=(10,10))

    plt.imshow(atlas.T)
```

```python
mandelbrot(120, -2.25,0.75, -1.5, 1.5, 1000)
```
